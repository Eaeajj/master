```python
import math
SECOND = 1_000_000

def find_max_n(t, f):
    if f == 'lg n':
        return 10**6 * t
    elif f == 'sqrt(n)':
        return (10**12 * t)**2
    elif f == 'n':
        return 10**6 * t
    elif f == 'n lg n':
        return (10**6 * t) // math.log10(10**6 * t)
    elif f == 'n^2':
        return math.isqrt(10**6 * t)
    elif f == 'n^3':
        return round(10**6 * t ** (1/3))
    elif f == '2^n':
        return math.log2(t)
    elif f == 'n!':
        n = 1
        fact = 1
        while fact < t:
            n += 1
            fact *= n
        return n
    else:
        return None

time_intervals = {
    'секунда': SECOND,
    'минута': 60 * SECOND,
    'час': 60 * 60 * SECOND,
    'день': 24 * 60 * 60 * SECOND,
    'месяц': 30 * 24 * 60 * 60 * SECOND,
    'год': 365 * 24 * 60 * 60 * SECOND,
    'век': 100 * 365 * 24 * 60 * 60 * SECOND
}

functions = ['lg n', 'sqrt(n)', 'n', 'n lg n', 'n^2', 'n^3', '2^n', 'n!']

for function in functions:
    values_n = []

    for interval_name, time_microseconds in time_intervals.items():
        max_n = find_max_n(time_microseconds, function)
        values_n.append(max_n)

    x = np.arange(len(time_intervals))
    width = 0.6

    plt.figure(figsize=(6, 6))

    plt.plot(x, values_n, "r--")

    plt.xlabel('Временной интервал')
    plt.yscale("log")
    plt.ylabel('Максимальное значение n')
    plt.title(f'Зависимость максимального значения n от времени для {function}')
    plt.xticks(x, time_intervals.keys())
    plt.grid(True)
    print(function)
    for i in values_n:
      print(f"{i:e}")
    print()


plt.show()
```

```python
import math

def find_max_n(t, f):
    if f == 'lg n':
        return 10**6 * t
    elif f == 'sqrt(n)':
        return (10**12 * t)**2
    elif f == 'n':
        return 10**6 * t
    elif f == 'n lg n':
        return (10**6 * t) // math.log10(10**6 * t)
    elif f == 'n^2':
        return math.isqrt(10**6 * t)
    elif f == 'n^3':
        return round(10**6 * t ** (1/3))
    elif f == '2^n':
        return math.log2(t)
    elif f == 'n!':
        n = 1
        fact = 1
        while fact < t:
            n += 1
            fact *= n
        return n
    else:
        return None

time_intervals = {
    'секунда': SECOND,
    'минута': 60 * SECOND,
    'час': 60 * 60 * SECOND,
    'день': 24 * 60 * 60 * SECOND,
    'месяц': 30 * 24 * 60 * 60 * SECOND,
    'год': 365 * 24 * 60 * 60 * SECOND,
    'век': 100 * 365 * 24 * 60 * 60 * SECOND
}

functions = ['2^n', 'n!']

for function in functions:
    values_n = []

    intervals = range(1, 1000, 100)
    for time_microseconds in intervals:
        max_n = find_max_n(time_microseconds, function)
        values_n.append(max_n)

    x = np.arange(len(intervals))
    plt.plot(x, values_n)
    plt.yscale("log")
    plt.xlabel('Временной интервал')
    plt.ylabel('Максимальное значение n')
    plt.grid(True)

plt.legend(functions)
plt.show()
```

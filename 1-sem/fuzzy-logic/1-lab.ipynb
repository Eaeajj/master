#### Задание 1: построить функции принадлежности

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# Задание 1: Треугольная и трапециевидная функции принадлежности
x = np.array([-1, -0.5, 0, 0.2, 1])

triangular_mf = fuzz.trimf(x, [-1, -0.5, 0.2])
trap_mf = fuzz.trapmf(x, [-1, -0.5, 0, 0.2])

# Построение графиков
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, triangular_mf, 'b', linewidth=2)
plt.title('Треугольная функция принадлежности')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, trap_mf, 'r', linewidth=2)
plt.title('Трапециевидная функция принадлежности')
plt.grid(True)

plt.tight_layout()
plt.show()

# Задание 2: Простая и двухсторонняя функции принадлежности Гаусса
gauss_simple_mf = fuzz.gaussmf(x, 0.5, 0)
gauss_two_sided_mf = fuzz.gaussmf(x, 0.2, 0.5)

# Построение графиков
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, gauss_simple_mf, 'g', linewidth=2)
plt.title('Простая функция принадлежности Гаусса')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, gauss_two_sided_mf, 'm', linewidth=2)
plt.title('Двухсторонняя функция принадлежности Гаусса')
plt.grid(True)

plt.tight_layout()
plt.show()

# Задание 4: Минимаксная интерпретация логических операторов
min_max_op = np.minimum(triangular_mf, trap_mf)

# Построение графика
plt.figure(figsize=(6, 4))
plt.plot(x, min_max_op, 'k', linewidth=2)
plt.title('Минимаксная интерпретация логических операторов')
plt.grid(True)
plt.show()

# Задание 5: Вероятностная интерпретация конъюнктивных и дизъюнктивных операторов
y = np.array([0.5, 0.03125, 0, 0.0008, 0.5])

# Вероятностная интерпретация конъюнктивных операторов (AND)
prob_and_op = np.minimum(triangular_mf, y)

# Вероятностная интерпретация дизъюнктивных операторов (OR)
prob_or_op = np.maximum(triangular_mf, y)

# Построение графиков
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, prob_and_op, 'r', linewidth=2)
plt.title('Вероятностная интерпретация конъюнктивных операторов (AND)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, prob_or_op, 'b', linewidth=2)
plt.title('Вероятностная интерпретация дизъюнктивных операторов (OR)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### Для одной из функций принадлежности привести пример использования

Рассмотрим пример использования одной из функций принадлежности, например, треугольной функции принадлежности. Предположим, у нас есть нечеткое множество, которое описывает температуру воздуха, и мы хотим определить, насколько воздух является "холодным".

```python
# Определение переменных
x = np.arange(-10, 30.1, 0.1)  # Диапазон температур
temperature = 15  # Текущая температура воздуха

# Треугольная функция принадлежности для "холодного" воздуха
# cold_air_mf = np.maximum(0, 1 - np.abs((x - 10) / 10))
cold_air_mf = fuzz.trimf(x, [0, 10, 20])

# Построение графика треугольной функции принадлежности
plt.figure(figsize=(10, 6))
plt.plot(x, cold_air_mf, 'b', linewidth=2, label='Функция принадлежности')
plt.title('Треугольная функция принадлежности для "холодного" воздуха')
plt.xlabel('Температура')
plt.ylabel('Принадлежность')

# Подсветим текущую температуру на графике
plt.plot(temperature, np.interp(temperature, x, cold_air_mf), 'ro', markersize=10, markerfacecolor='r', label='Текущая температура')
plt.legend(loc='best')
plt.grid(True)
plt.show()

# Оценка степени принадлежности текущей температуры "холодному" воздуху
membership_degree = np.interp(temperature, x, cold_air_mf)
print(f'Принадлежность текущей температуры "холодному" воздуху: {membership_degree}')
```

В этом примере мы определяем треугольную функцию принадлежности для "холодного" воздуха, строим ее график и подсвечиваем текущую температуру на графике. Затем мы оцениваем степень принадлежности текущей температуры к "холодному" воздуху.

### 6. Построить графики, иллюстрирующие операции объединения и пересечения (двумя методами), а также операцию дополнения. 

```python
# Определение переменных
x = np.arange(0, 10.1, 0.1)  # Диапазон уровня заполненности бака

# Треугольные функции принадлежности для "высокого" и "низкого" уровней воды
# high_level_mf = np.maximum(0, 1 - np.abs((x - 8) / 2))
# low_level_mf = np.maximum(0, 1 - np.abs((x - 2) / 2))
high_level_mf = fuzz.trimf(x, [6, 8, 10])
low_level_mf = fuzz.trimf(x, [0, 2, 4])

# Построение графиков треугольных функций принадлежности
plt.figure(figsize=(10, 12))

plt.subplot(3, 1, 1)
plt.plot(x, high_level_mf, 'b', linewidth=2)
plt.title('Функция принадлежности для "высокого" уровня воды')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x, low_level_mf, 'r', linewidth=2)
plt.title('Функция принадлежности для "низкого" уровня воды')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

# Операция объединения (OR)
union_result = np.maximum(high_level_mf, low_level_mf)
plt.subplot(3, 1, 3)
plt.plot(x, union_result, 'g', linewidth=2)
plt.title('Операция объединения (OR)')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

# Регулировка расстояния между графиками
plt.subplots_adjust(hspace=0.3)

# Операция пересечения (AND) - Минимум
intersection_result_min = np.minimum(high_level_mf, low_level_mf)

# Операция пересечения (AND) - Произведение
intersection_result_prod = high_level_mf * low_level_mf

# Построение графиков для операций пересечения (AND)
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(x, intersection_result_min, 'm', linewidth=2)
plt.title('Операция пересечения (AND) - Минимум')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, intersection_result_prod, 'c', linewidth=2)
plt.title('Операция пересечения (AND) - Произведение')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

# Регулировка расстояния между графиками
plt.subplots_adjust(hspace=0.3)

# Операция дополнения
complement_high_level = 1 - high_level_mf
complement_low_level = 1 - low_level_mf

# Построение графиков для операции дополнения
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(x, complement_high_level, 'r', linewidth=2)
plt.title('Операция дополнения для "высокого" уровня воды')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, complement_low_level, 'b', linewidth=2)
plt.title('Операция дополнения для "низкого" уровня воды')
plt.xlabel('Уровень заполненности')
plt.ylabel('Принадлежность')
plt.grid(True)

# Регулировка расстояния между графиками
plt.subplots_adjust(hspace=0.3)

plt.show()
```

#### 7. Сконструировать нечёткую систему, отображающую зависимость между переменными x и у, заданную с помощью табл.2.

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# Заданные значения аргумента и функции
x = np.array([-1, -0.5, 0, 0.2, 1])
y = np.array([1, 0.25, 0, 0.4, 1])

# Построение графиков в одном окне
plt.figure(figsize=(8, 10))

# График зависимости между переменными x и y
plt.subplot(2, 1, 1)
plt.plot(x, y, 'bo-', linewidth=2)
plt.title('Зависимость между переменными x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# График нечеткой системы
plt.subplot(2, 1, 2)

# Определение функций принадлежности для нечетких множеств
mf1 = fuzz.trapmf(x, [-1, -0.5, 0, 0.2])
mf2 = fuzz.trimf(x, [-0.5, 0, 0.2])
mf3 = fuzz.trapmf(x, [0, 0.2, 1, 1])

# Нечеткая инференция
inference_result_fuzz = np.maximum(
  fuzz.interp_membership(x, mf1, x),
  np.maximum(
    fuzz.interp_membership(x, mf2, x),
    fuzz.interp_membership(x, mf3, x)
    )
  )

plt.plot(x, inference_result_fuzz, 'r', linewidth=2)
plt.title('Нечеткая система')
plt.xlabel('x')
plt.ylabel('Нечеткая инференция (y)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

2. Чем различаются нечёткая и лингвистическая переменные?  
Нечёткая переменная:
$$ < 𝑎,𝑋,𝐴 >, $$
$$ ⟨теплая,\{0,+70\},\{23,33\}⟩ $$
Лингвистическая переменная:
$$< 𝑥, 𝑇,𝑈, 𝐺, 𝑀 >,$$
$$⟨Температура воды,Т,U,G,M⟩$$

где:
  Т - базовое терм-множество (например, {холодная, прохладная, теплая, горячая});  
  U - область определения нечётких переменных (допустим, от 0 до 100 градусов по Цельсию);  
  G - синтаксическая процедура формирования новых значений (например, правила для определения "очень холодной" или "умеренно горячей");  
  M - семантическая процедура, которая ставит в соответствие каждому значению осмысленное содержание (например, "холодная" соответствует температуре 0-20 градусов "теплая" - 20-40 градусов).  

Лингвистические переменные являются нечеткими переменными, но не все нечеткие переменные являются лингвистическими переменными. Ключевое различие заключается в том, как представлены степени достоверности: для нечетких переменных это делается с помощью функций принадлежности, а для лингвистических переменных - с помощью лингвистических терминов.

3. Что такое нечёткое отношение?
$$
R \subset X \times X
$$

$$
\mu_R: X \times X \to [0, 1]
$$

$$
\mu_R(x, y) \text{ - субъективная мера выполнения отношения } xRy
$$  

Словами: Нечеткое отношение R на множестве X - это какое-то нечеткое правило или условие, которое определяет, какие элементы из множества X связаны между собой.
То есть, вместо того, чтобы точно говорить, что "элемент x в отношении R к элементу y", мы используем нечеткую функцию принадлежности μR(x, y), которая говорит о том, насколько сильно элементы x и y связаны в этом отношении. Эта функция принимает значения от 0 до 1, где 0 означает отсутствие связи, а 1 - полная связь.  

4. Какими свойствами обладают нечёткие отношения?  
- **Рефлексивность.** Нечёткое отношение $R$ рефлексивно на $X$, если $\mu_R(x, x) = 1$ для всех $x \in X$.
- **Антирефлексивность.** Нечёткое отношение $R$ антирефлексивно на $X$, если $\mu_R(x, x) = 0$ для всех $x \in X$.

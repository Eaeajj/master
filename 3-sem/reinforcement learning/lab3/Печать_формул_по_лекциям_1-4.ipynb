#–ú–Ω–æ–≥–æ—Ä—É–∫–∏–µ –±–∞–Ω–¥–∏—Ç—ã

$$q^*(a) = ùîº[R_t | A_t = a]$$

$A_t$ - –¥–µ–π—Å—Ç–≤–∏–µ, –≤—ã–±—Ä–∞–Ω–Ω–æ–µ –Ω–∞ –≤—Ä–µ–º–µ–Ω–Ω–æ–º —à–∞–≥–µ t, $R_t$ ‚Äì —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –µ–º—É –≤–æ–∑–Ω–∞–≥—Ä–∞–∂–¥–µ–Ω–∏–µ, $q^*(a)$ - –∏—Å—Ç–∏–Ω–Ω–∞—è —Ü–µ–Ω–Ω–æ—Å—Ç—å –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–≥–æ –¥–µ–π—Å—Ç–≤–∏—è $a$ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ –µ–≥–æ –≤—ã–±–æ—Ä–∞.



$$Q_t(a) = \frac{\sum_{n=1}^{t-1}R_n*ùüô_{A_i=a}}{\sum_{n=1}^{t-1}  ùüô_{A_i=a}},$$
$$Q_n = \frac{R_1+R_2+...+R_{n-1}}{n-1}$$

$Q_t(a)$ –∏ $Q_n$ - –æ—Ü–µ–Ω–∫–∞ —Ü–µ–Ω–Ω–æ—Å—Ç–∏ –¥–µ–π—Å—Ç–≤–∏—è, $ùüô_{A_i=a}$ ($ùüô_{predicate}$ –≤ –æ–±—â–µ–π —Ñ–æ—Ä–º–µ) –æ–±–æ–∑–Ω–∞—á–∞–µ—Ç —Å–ª—É—á–∞–π–Ω—É—é –≤–µ–ª–∏—á–∏–Ω—É, —Ä–∞–≤–Ω—É—é 1, –µ—Å–ª–∏ –ø—Ä–µ–¥–∏–∫–∞—Ç predicate —Ä–∞–≤–µ–Ω true, –∏ 0 –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

$$A_t = argmax_a(Q_t(a)),$$

–ñ–∞–¥–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ, $argmax_a$ –æ–±–æ–∑–Ω–∞—á–∞–µ—Ç –¥–µ–π—Å—Ç–≤–∏–µ, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ —Å–ª–µ–¥—É—é—â–µ–µ –¥–∞–ª–µ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –º–∞–∫—Å–∏–º—É–º.

$$Q_{n+1} = \frac{1}{n}\sum_{i=1}^{n}R_i = \frac{1}{n}(R_n+\sum_{i=1}^{n-1}R_i)= \frac{1}{n}(R_n+(n-1)\frac{1}{n-1}\sum_{i=1}^{n-1}R_i) = $$
$$ = \frac{1}{n}(R_n+(n-1)Q_n) = \frac{1}{n}(R_n+nQ_n-Q_n) = Q_n+\frac{1}{n}[R_n-Q_n]$$

–ò–Ω–∫—Ä–µ–º–µ–Ω—Ç–Ω—ã–π –≤–∞—Ä–∏–∞–Ω—Ç –æ—Ü–µ–Ω–∫–∏ —Ü–µ–Ω–Ω–æ—Å—Ç–∏ –¥–µ–π—Å—Ç–≤–∏—è.

–û–±–æ–±—â–∞—è —ç—Ç—É —Ñ–æ—Ä–º—É–ª—É, –ø–æ–ª—É—á–∞–µ–º:
$$–ù–æ–≤–∞—è–û—Ü–µ–Ω–∫–∞ ‚üµ –°—Ç–∞—Ä–∞—è–û—Ü–µ–Ω–∫–∞ + –†–∞–∑–º–µ—Ä–®–∞–≥–∞[–¶–µ–ª—å - –°—Ç–∞—Ä–∞—è–û—Ü–µ–Ω–∫–∞]$$
$$Q_{n+1} = Q_n+\alpha[R_n-Q_n]$$

# –ö–æ–Ω–µ—á–Ω—ã–µ –º–∞—Ä–∫–æ–≤—Å–∫–∏–µ –ø—Ä–æ—Ü–µ—Å—Å—ã –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏–π - –∫–æ–Ω–µ—á–Ω—ã–µ –ú–ü–ü–†

–ü—Ä–∏ –≤–≤–æ–¥–µ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –º–æ–º–µ–Ω—Ç–æ–≤ –≤—Ä–µ–º–µ–Ω–∏ t = 0, 1, 2, ..., –ø–æ–ª—É—á–∏–º —Å–ª–µ–¥—É—â—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–π: $S_0, A_0, R_1, S_1, A_1, R_2, S_2, A_2, R_3, S_3, A_3,...$.

–í–Ω–∞—á–∞–ª–µ –∞–≥–µ–Ω—Ç –ø–æ–ª—É—á–∞–µ—Ç —Å–æ—Å—Ç–æ—è–Ω–∏–µ —Å—Ä–µ–¥—ã $S_t ‚àà ùíÆ$, –∑–∞—Ç–µ–º –≤—ã–±–∏—Ä–∞–µ—Ç –¥–µ–π—Å—Ç–≤–∏–µ $A_t ‚àà ùíú(S_t)$, –Ω–∞ —Å–ª–µ–¥—É—é—â–µ–º –≤—Ä–µ–º–µ–Ω–Ω–æ–º —à–∞–≥–µ, –æ—Ç—á–∞—Å—Ç–∏ –∫–∞–∫ –ø–æ—Å–ª–µ–¥—Å—Ç–≤–∏–µ —Å–≤–æ–µ–≥–æ –≤—ã–±–æ—Ä–∞ $A_t$, –æ–Ω –ø–æ–ª—É—á–∞–µ—Ç —á–∏—Å–ª–æ–≤—É—é –Ω–∞–≥—Ä–∞–¥—É $R_{t+1} ‚àà \mathscr{R}$ –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –Ω–æ–≤–æ–º —Å–æ—Å—Ç–æ—è–Ω–∏–∏ —Å—Ä–µ–¥—ã $S_{t+1}$.

---
–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:

$\mathscr{S}$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ nonterminal (–Ω–µ–∫–æ–Ω–µ—á–Ω—ã—Ö) —Å–æ—Å—Ç–æ—è–Ω–∏–π

$\mathscr{A}(s)$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤—Å–µ—Ö –¥–µ–π—Å—Ç–≤–∏–π, –¥–æ—Å—Ç—É–ø–Ω—ã—Ö –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–∏ s

$\mathscr{R}$ - –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤—Å–µ—Ö –Ω–∞–≥—Ä–∞–¥

ùúã - –ø–æ–ª–∏—Ç–∏–∫–∞(–ø—Ä–∞–≤–∏–ª–æ –≤—ã–±–æ—Ä–∞ –¥–µ–π—Å—Ç–≤–∏–π)

$s$ - —Å–æ—Å—Ç–æ—è–Ω–∏–µ

$s'$ - –Ω–æ–≤–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ

$a$ - –¥–µ–π—Å—Ç–≤–∏–µ

$r$ - –Ω–∞–≥—Ä–∞–¥–∞

$p$ - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–µ—Ä–µ—Ö–æ–¥–∞ –∏–∑ $s$ –≤ $s'$

$T$ - –ø–æ—Å–ª–µ–¥–Ω–∏–π –≤—Ä–µ–º–µ–Ω–Ω–æ–π —à–∞–≥ –≤ —ç–ø–∏–∑–æ–¥–µ

terminal - –∫–æ–Ω–µ—á–Ω–æ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –≤ —Å—Ä–µ–¥–µ (–ø–æ–ø–∞–¥–∞—è –≤ –Ω–µ–≥–æ –∞–≥–µ–Ω—Ç –∑–∞–∫–∞–Ω—á–∏–≤–∞–µ—Ç —ç–ø–∏–∑–æ–¥)

---
$Œ≥$ - –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –æ–±–µ—Å—Ü–µ–Ω–∏–≤–∞–Ω–∏—è

$Œ±$ - –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –æ–±—É—á–µ–Ω–∏—è –∏–ª–∏ —Ä–∞–∑–º–µ—Ä —à–∞–≥–∞, –≤ —Ñ–æ—Ä–º—É–ª–∞—Ö –º–Ω–æ–≥–æ—Ä—É–∫–∏—Ö –±–∞–Ω–¥–∏—Ç–æ–≤ $Œ±=\frac{1}{n}$

---
$v, v_œÄ(s), V, V_i, V(S), V_i(S), V(S_j), V_i(S_j)$ - –æ—Ü–µ–Ω–∫–∞ —Ü–µ–Ω–Ω–æ—Å—Ç–∏ —Å–æ—Å—Ç–æ—è–Ω–∏—è

$q, q_œÄ(s,a), Q, Q_i, Q(S,A), Q_i(S,A), Q(S_j, A_j), Q_i(S_j, A_j)$ - –æ—Ü–µ–Ω–∫–∞ —Ü–µ–Ω–Ω–æ—Å—Ç–∏ –¥–µ–π—Å—Ç–≤–∏—è

$v^*, v^*(s)$ - –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è (–∏—Å—Ç–∏–Ω–Ω–∞—è) —Ü–µ–Ω–Ω–æ—Å—Ç—å —Å–æ—Å—Ç–æ—è–Ω–∏—è

$q^*, q^*(a), q^*(s,a)$ - –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è (–∏—Å—Ç–∏–Ω–Ω–∞—è) —Ü–µ–Ω–Ω–æ—Å—Ç—å –¥–µ–π—Å—Ç–≤–∏—è

---


–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤–æ–∑–Ω–∏–∫–Ω–æ–≤–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π $s',r,s,a$ –≤ –º–æ–º–µ–Ω—Ç $t$:

$\forall s‚àà\mathscr{S}, a‚àà\mathscr{A}(s)$ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è:

$$\sum_{s'‚àà\mathscr{S}}\sum_{r‚àà\mathscr{R}} p(s',r|s,a) = 1$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–µ—Ä–µ—Ö–æ–¥–∞ —Å–æ—Å—Ç–æ—è–Ω–∏–π:

$$p(s'|s,a) = Pr{S_t = s'|S_{t-1} = s, A_{t-1} = a}  = \sum_{r‚àà\mathscr{R}} p(s',r|s,a)$$


–û–∂–∏–¥–∞–µ–º—ã–µ –≤–æ–∑–Ω–∞–≥—Ä–∞–∂–¥–µ–Ω–∏—è –¥–ª—è –ø–∞—Ä —Å–æ—Å—Ç–æ—è–Ω–∏–µ‚Äì–¥–µ–π—Å—Ç–≤–∏–µ:

$$r(s,a) = \mathbb{E}[R_t|S_{t-1}=s,A_{t-1}=a] = \sum_{r‚àà‚Ñõ}r\sum_{s'‚àà\mathscr{S}}p(s',r|s,a)$$

–û–∂–∏–¥–∞–µ–º—ã–µ –≤–æ–∑–Ω–∞–≥—Ä–∞–∂–¥–µ–Ω–∏—è –¥–ª—è —Ç—Ä–æ–µ–∫ —Å–æ—Å—Ç–æ—è–Ω–∏–µ‚Äì–¥–µ–π—Å—Ç–≤–∏–µ‚Äì—Å–ª–µ–¥—É—é—â–µ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ:

$$r(s,a,s') = \mathbb{E}[R_t|S_{t-1}=s,A_{t-1}=a,S_t = s'] = \sum_{r‚àà‚Ñõ}r\frac{p(s',r|s,a)}{p(s'|s,a)}$$

–ß–∞—â–µ –≤—Å–µ–≥–æ –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å $p(s',r|s,a)$.

–û–∂–∏–¥–∞–µ–º—ã–π –¥–æ—Ö–æ–¥:

$$G_t = R_{t+1} + Œ≥R_{t+2} + Œ≥^2R_{t+3} + ... + Œ≥^{T-1}R_{T} = \sum_{k=t+1}^{T}Œ≥^{k-t-1}R_{k} = R_{t+1} + Œ≥(R_{t+2} + Œ≥R_{t+3} + Œ≥^2R_{t+4} + ...) = R_{t+1} + Œ≥G_{t+1},$$

$T$ ‚Äì –ø–æ—Å–ª–µ–¥–Ω–∏–π –≤—Ä–µ–º–µ–Ω–Ω–æ–π —à–∞–≥ –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–µ terminal, $Œ≥$ - –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –æ–±–µ—Å—Ü–µ–Ω–∏–≤–∞–Ω–∏—è, $0\leq Œ≥ \leq 1$.


$G_T = 0$

–ü—Ä–∏ $T = \infty$, $G_t$ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Å—É–º–º–æ–π –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã—Ö —Å–ª–∞–≥–∞–µ–º—ã—Ö. –ü—Ä–∏ $R = const = 1$ –¥–æ—Ö–æ–¥ –±—É–¥–µ—Ç:
$$G_t = \sum_{k=1}^{‚àû}{Œ≥^{k}} = \frac{1}{(1-Œ≥)}.$$

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –ë–µ–ª–ª–º–∞–Ω–∞ –¥–ª—è $v_œÄ(s)$:

$$v_œÄ(s) = \mathbb{E}_œÄ[G_t|S_t=s] = \mathbb{E}_œÄ[\sum_{k=0}^{\infty}Œ≥^{k}R_{t+k+1}|S_t=s] = \mathbb{E}_œÄ[R_{t+1}+Œ≥G_{t+1}|S_t=s] = $$
$$ = \sum_{a}œÄ(a|s)\sum_{s',r}p(s',r|s,a)\Big[r+Œ≥\mathbb{E}_œÄ[G_{t+1}|S_{t+1}=s']\Big] = $$
$$ = \sum_{a}œÄ(a|s)\sum_{s',r}p(s',r|s,a)[r+Œ≥v_œÄ(s')] = \sum_{a}œÄ(a|s)q_œÄ(s,a)$$

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –ë–µ–ª–ª–º–∞–Ω–∞ –¥–ª—è $q_œÄ(s,a)$:
$$q_œÄ(s,a) = \mathbb{E}_œÄ[G_t|S_t=s, A_t=a]$$  
$$=  \mathbb{E}_œÄ[R_{t+1}+Œ≥G_{t+1}|S_t=s, A_t=a] = \sum_{s',r}p(s',r|s,a)[r+Œ≥\sum_{a'}œÄ(a'|s')q_œÄ(s',a')]$$

–§—É–Ω–∫—Ü–∏—è –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–∞ –¥–µ–π—Å—Ç–≤–∏—è –∏–ª–∏ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–∞, A-—Ñ—É–Ω–∫—Ü–∏—è:
$$a_œÄ(s, a) = q_œÄ(s, a) - v_œÄ(s).$$

–°—Ç—Ä–∞—Ç–µ–≥–∏—è $œÄ$ –±–æ–ª—å—à–µ (–ª—É—á—à–µ) –∏–ª–∏ —Ä–∞–≤–Ω–∞ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ $œÄ'$, —Ç–æ–≥–¥–∞ –∏ —Ç–æ–ª—å–∫–æ —Ç–æ–≥–¥–∞, –∫–æ–≥–¥–∞ $v_œÄ(s) ‚â• v_{œÄ'}(s)$ –¥–ª—è –≤—Å–µ—Ö $s ‚àà ùíÆ$. –í—Å–µ–≥–¥–∞ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ö–æ—Ç—è –±—ã –æ–¥–Ω–∞ —Å—Ç—Ä–∞—Ç–µ–≥–∏—è, –±–æ–ª—å—à–∞—è –∏–ª–∏ —Ä–∞–≤–Ω–∞—è –≤—Å–µ–º –æ—Å—Ç–∞–ª—å–Ω—ã–º —Å—Ç—Ä–∞—Ç–µ–≥–∏—è–º - –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è —Å—Ç—Ä–∞—Ç–µ–≥–∏—è. –£ –≤—Å–µ—Ö –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö —Å—Ç—Ä–∞—Ç–µ–≥–∏–π –æ–¥–Ω–∞ –∏ —Ç–∞ –∂–µ —Ñ—É–Ω–∫—Ü–∏—è —Ü–µ–Ω–Ω–æ—Å—Ç–∏ —Å–æ—Å—Ç–æ—è–Ω–∏–π, –∫–æ—Ç–æ—Ä–∞—è –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π —Ü–µ–Ω–Ω–æ—Å—Ç–∏ —Å–æ—Å—Ç–æ—è–Ω–∏–π:

$$v^*(s) = max_œÄ v_œÄ(s) \forall ‚àà\mathscr{S}.$$

–£ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö —Å—Ç—Ä–∞—Ç–µ–≥–∏–π —Ç–∞–∫–∂–µ –æ–¥–Ω–∞ –∏ —Ç–∞ –∂–µ —Ñ—É–Ω–∫—Ü–∏—è —Ü–µ–Ω–Ω–æ—Å—Ç–∏ –¥–µ–π—Å—Ç–≤–∏–π:
$\forall s‚àà\mathscr{S}, a‚àà\mathscr{A}(s)$
$$q^*(s,a) = max_œÄ q_œÄ(s,a) = \mathbb{E}[R_{t+1}+Œ≥v^*(S_{t+1})|S_t=s,A_t=a].$$

–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ—Å—Ç–∏ –ë–µ–ª–ª–º–∞–Ω–∞:

$$v^*(s) = max_{a‚àà\mathscr{A}(s)} q_{œÄ^*}(s,a) = max_{a}\mathbb{E}_{œÄ^*}[G_t|S_t=s,A_t=a] = max_{a}\mathbb{E}_{œÄ^*}[R_{t+1}+Œ≥G_{t+1}|S_t=s,A_t=a] = max_{a}\mathbb{E}[R_{t+1}+Œ≥v^*(S_{t+1})|S_t=s,A_t=a] = max_{a}\sum_{s',r}p(s',r|s,a)[r+Œ≥v^*(s')]$$


–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ—Å—Ç–∏ –ë–µ–ª–ª–º–∞–Ω–∞ –¥–ª—è $q^*$:

(12)$$q^*(s,a) = \mathbb{E}[R_{t+1}+Œ≥max_{a'}q^*(S_{t+1},a')|S_t=s,A_t=a] = \sum_{s',r}p(s',r|s,a)[r+Œ≥max_{a'}q^*(s',a')]$$


#–î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ - –î–ü

–ò–Ω–∫—Ä–µ–º–µ–Ω—Ç–Ω–∞—è —Ñ–æ—Ä–º–∞ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –ë–µ–ª–ª–º–∞–Ω–∞ –¥–ª—è $v(s)$:

$$v_{k+1}(s) = \mathbb{E}_œÄ[R_{t+1}+Œ≥v_k(S_{t+1})|S_t = s] = \sum_{a}œÄ(a|s)\sum_{s',r}p(s',r|s,a)[r+Œ≥v_k(s')]$$

–¢–µ–æ—Ä–µ–º–∞ –æ–± —É–ª—É—á—à–µ–Ω–∏–∏ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏. –ü—É—Å—Ç—å $œÄ$ –∏ $œÄ'$ ‚Äì –ª—é–±–∞—è –ø–∞—Ä–∞ –¥–µ—Ç–µ—Ä–º–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Å—Ç—Ä–∞—Ç–µ–≥–∏–π —Ç–∞–∫–∞—è, —á—Ç–æ $\forall s ‚àà ùíÆ:$

$$q_œÄ(s, œÄ'(s)) \geq v_œÄ(s).$$

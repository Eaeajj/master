# Конечные марковские процессы принятия решений - конечные МППР

>МППР – классическая формализация последовательного принятия решений, когда от действий зависит не только немедленное вознаграждение, но и последующие ситуации, или состояния, а через них и будущие вознаграждения. Раньше мы такую связь не рассматривали, т.е. мы работали с обычными марковскими процессами - «будущее» процесса не зависит от «прошлого» при известном «настоящем».


В предыдущей методичке мы смотрели на так называемые марковские процессы, где решения не влияли друг на друга и на среду. Теперь перейдем к более реальным примерам, где каждое действие будет влиять не только на состояние среды, но и на последующие варианты действий и, конечно, на награду, т.е. траектория будет состоять уже не из одного состояния и действия, а из нескольких. За это отвечают марковские процессы принятия решений.

## Монополия

hasbro_00009e88_images_1410174772-min.jpg

Играют двое: мы и наш противник. Обозначим нашу фишку на поле треугольником, а нашего противника - квадратом. Передвигаться можно на расстояние от 1 до 6 клеток игрового поля.

На каждой клетке-улице игрового поля, помеченных ценой и не купленных, есть два варианта: купить улицу или не покупать. Если игрок попадает на улицу, которую купил другой игрок, то он должен заплатить этому игроку 20% от стоимости этой улицы.

Кроме этого у нас есть четыре клетки Шанс. На них рандомно выпадают карточки: получить приз $10$ рублей, оплатить штраф в банк $5$ рублей, пропустить 1 ход. И три поля с пропуском разного количества ходов.

Каждый игрок начинает со $100$ рублями. Условия проигрыша: потеря всех денег.

Теперь перед нами стоит вопрос: какую стратегию выбрать? Можно воспользоваться самой простой: купить все, что попадется на пути! По ощущениям ведь чем больше купим сами, тем меньше достанется противнику, еще и заставим его платить нам 20% с каждой нашей улицы. Однако, учитывая ограничение начального капитала в $100$ рублей, эта стратегия вряд ли окупится в долгосрочной перспективе.

image.png

Представим, что находимся на предпоследней клетке перед завершением круга. После круга, пройденного согласно такой стратегии, у нас остался $41$ рубль. Перед нами стоит вопрос: купить улицу, на которой мы находимся или нет? Она стоит $40$ рублей.

Если мы купим эту улицу, то на следующем ходе наши шансы проиграть значительно вырастут: 4 поля из 6 для нас плохие.

Таким образом, наша стратегия скупать все должна перерасти в нечно большее: более оптимальную стратегию. Чтобы ее найти, нужно разобраться с вероятностями и исследованием среды. В этом помогут марковские цепи и марковские процессы принятия решений.



## Основные определения

diplom-Страница-2.jpg

При вводе дискретной последовательности моментов времени t = 0, 1, 2, ..., получим следущую последовательность взаимодействий: $S_0, A_0, R_1, S_1, A_1, R_2, S_2, A_2, R_3, S_3, A_3,...$, т.к. вначале агент получает состояние среды $S_t$ ∈ $𝒮$, затем выбирает действие $A_t$ ∈ $𝒜(S_t)$, на следующем временном шаге, отчасти как последствие своего выбора $A_t$, он получает числовую награду $R_{t+1} ∈ \mathscr{R}$ и информацию о новом состоянии среды $S_{t+1}$.

---
Дополнительные обозначения:

$\mathscr{S}$ - множество nonterminal (неконечных) состояний

$\mathscr{S}^+$ - множество всех состояний, включая terminal (конечное состояние)

$\mathscr{A}$(s) - множество всех действий, доступных в состоянии s

$\mathscr{R}$ - множество всех наград

𝜋 - политика(правило выбора действий)

$s$ - состояние

$s'$ - новое состояние

$a$ - действие

$r$ - награда

$p$ - вероятность перехода из $s$ в $s'$

---

Графически **<font color='orange'>марковская цепь</font>** будет выглядеть, как последовательность некоторых состояний одного процесса с некоторыми вероятностями перехода в них. Все возможные состояния процесса образуют пространство состояний, оно конечно. Последовательность состояний процесса образует цепочку, отсюда и название. **<font color='orange'>Важно:</font>** цепь Маркова требует, чтобы вероятность перехода в следующее состояние зависела **<font color='orange'>только от текущего состояния</font>**.

Добавим к нашей графической схеме еще одну деталь - награду за переход из одного состояния в другое. Это **<font color='orange'>марковский процесс вознаграждения</font>**. Если в предыдущем варианте главный вопрос был "Какова вероятность перехода?", то в данном случае "Какова вероятность получить x награды после n шагов?".

Существует много вариаций таких диаграмм.

А теперь, опираясь на эти данные, попробуем выяснить, какова вероятность нашего проигрыша, если мы все же купим сейчас улицу за 40 рублей. Итак, две коричневые улицы Vediterranean avenue и Baltic avenue, один исход у шанса, Vermont avenue:

$$\frac{1}{6}+\frac{1}{6}+\frac{1}{18}+\frac{1}{6} = \frac{5}{9}$$

Это означает, что у нас больше 50% вероятность проиграть. Следовательно, вариант с покупкой улицы очень рискованный.
Теперь перейдем к МППР, где текущие решения зависят от предыдущих.

### Обсуждение


1.   Какова вероятность при такой диаграмме переходов закончить отпуск с суммарной наградой $r = 0$?
Screenshot_20240515_130924.png

В конечном МППР множества состояний, действий и вознаграждений ($𝒮, 𝒜$ и $ℛ$) содержат конечное число элементов. Тогда для случайных $s' ∈ 𝒮$ и $r ∈ ℛ$, определена **<font color='orange'>вероятность возникновения этих значений в момент $t$</font>**, при условии известных значений предшествующего состояния $s$ и действия $a$:

$p(s',r|s,a) = $ Pr{$S_t = s', R_t = r|S_{t-1} = s, A_{t-1} = a$}

$p$ задает распределение вероятностей для каждого выбора $s$ и $a$, полностью характеризуя динамику среды, т.е.

$\forall s∈\mathscr{S}, a∈\mathscr{A}(s)$ выполняется:

(1) $$\sum_{s'∈\mathscr{S}}\sum_{r∈\mathscr{R}} p(s',r|s,a) = 1$$

То есть вероятность каждого возможного значения $S_t$ и $R_t$ зависит только от непосредственно предшествующего состояния $S_{t–1}$ и действия $A_{t–1}$, и не зависит от всех предыдущих состояний и действий $t-2, t-3, ...$ Лучше считать это ограничением не на процесс принятия решений, а на состояние.

Таким образом, состояние должно включать информацию обо всех аспектах прошлых взаимодействий агента со средой, которые существенны для будущего. Если это так, то говорят, что состояние обладает **<font color='orange'>марковским свойством</font>**.

Зная функцию динамики $p$ с четырьмя аргументами, можно вычислить все, что
нужно знать о среде, например, **<font color='orange'>вероятность
 перехода состояний</font>**:

(2) $p(s'|s,a) = $ Pr{$S_t = s'|S_{t-1} = s, A_{t-1} = a$} $ = \sum_{r∈\mathscr{R}} p(s',r|s,a)$
>вероятность перехода в $s'$ из $s$ после выбора действия $a$.

**<font color='orange'>Ожидаемые вознаграждения</font>** для пар состояние–действие:

(3) $r(s,a) = \mathbb{E}[R_t|S_{t-1}=s,A_{t-1}=a] = \sum_{r∈ℛ}r\sum_{s'∈\mathscr{S}}p(s',r|s,a)$
>ожидаемое немедленное вознаграждение в $s$ после выбора действия $a$.

**<font color='orange'>Ожидаемые вознаграждения</font>** для троек состояние–действие–следующее состояние:

(4) $r(s,a,s') = \mathbb{E}[R_t|S_{t-1}=s,A_{t-1}=a,S_t = s'] = \sum_{r∈ℛ}r\frac{p(s',r|s,a)}{p(s'|s,a)}$
>ожидаемое немедленное вознаграждение при переходе из $s$ в $s'$ после выбора действия $a$.

Чаще всего мы будем использовать $p(s',r|s,a)$.

## Граница между средой и агентом

>Структура MDP представляет собой значительную абстракцию проблемы целенаправленного обучения на основе взаимодействия. Каковы бы ни были детали механизмов восприятия, памяти и управления и какой бы цели агент ни стремился достичь, любая задача обучения целенаправленному поведению может быть сведена к трем сигналам, которыми агент обменивается с окружающей средой: действия, состояния и награды.

Общее правило, которому мы следуем, таково: **<font color='orange'>все, что не может быть произвольным образом изменено агентом, считается находящимся вне его, а значит, частью окружающей среды</font>**. При этом среда может сообщать агенту и эти данные. Например, агент часто знает о том, как вознаграждения вычисляются в виде функций от его действий и состояний, в которых они предприняты. Но мы всегда рассматриваем процесс вычисления вознаграждения как внешний по отношению к агенту (происходящий в среде), потому что он определяет стоящую перед агентом задачу, и, следовательно, агент не способен изменять его произвольным образом.

Граница между агентом и средой определяет пределы возможностей **<font color='orange'>абсолютного контроля</font>** со стороны агента, но не ограничивает его знания. Кроме этого, граница может проходить в разных местах в зависимости от целей поставленной задачи.

### Пример №1. Биореактор.

Рассмотрим применение обучения с подкреплением к определению температур и скоростей перемешивания в биореакторе в последовательные моменты времени. В таком приложении **<font color='orange'>действиями</font>** могут быть целевая температура и скорость перемешивания, передаваемые низкоуровневым управляющим системам, которые, в свою очередь, активируют нагревательные элементы и моторы для достижения заданных показателей. **<font color='orange'>Состояниями</font>** будут показания термопары и других датчиков, возможно, подвергнутые фильтрации и поступающие с задержкой, а также символические входные данные, описывающие ингредиенты в чане и конечное вещество. **<font color='orange'>Вознаграждениями</font>** могут быть дискретные результаты измерения скорости производства полезного химиката биореактором.

### Пример №2. Перегрузочный робот.

Теперь попробуем примененить обучение с подкреплением к управлению движением роботизированной руки, выполняющей повторяющуюся задачу типа «взять и положить». Если мы хотим обучить робота быстрым и плавным движениям, то обучаемый агент должен напрямую управлять моторами и с небольшой задержкой получать информацию о текущих позициях и скоростях механических сочленений. **<font color='orange'>Действиями</font>** в этом случае могут быть напряжения, подаваемые на каждый мотор в каждом сочленении, а **<font color='orange'>состояниями</font>** – последние данные об углах и скоростях сочленений. **<font color='orange'>Вознаграждение</font>** может быть равно +1 для каждого успешно перемещенного объекта. Чтобы поощрить плавные движения, на каждом шаге можно выдавать **<font color='orange'>небольшое отрицательное вознаграждение</font>**, вычисляемое как функция от неравномерности движения в соседние моменты времени.

### Пример №3. Видеоигры.

Один из вариантов рассмотрения видеоигр в обучение с подкреплением: агент - игрок, среда - информация с экрана. **<font color='orange'>Действиями</font>** в этом случае становятся клики мыши, клавиши клавиатуры или стики и кнопки на джойстике, а **<font color='orange'>состояниями</font>** - переработанные скриншоты или видео экрана с игрой. Иногда вместо скринов напрямую подается какая-то конкретная информация, но в начале это все равно обрабатывется с экрана или файлов игры. **<font color='orange'>Вознаграждение</font>** может быть равно собранным монетам, уничтоженым врагам или чему-то еще, выраженному в числовом эквиваленте. То, сколько все это будет стоит в итоге решает разработчик алгоритма обучения, корректируя таким образом важность каких-либо задач.

### Пример №4. Робот-пылесос

Подвижный робот собирает мусор в офисе. У него имеются датчики для обнаружения мусора. Питается робот от перезаряжаемого аккумулятора. В системе управления роботом имеются компоненты для интерпретации получаемой от датчиков информации, для навигации и для управления сбором мусора. Решения верхнего уровня о том, как искать мусор, принимаются обучающимся с подкреплением агентом на основе информации о текущем заряде аккумулятора.

image.png

Для простоты предположим, что различается всего два уровня, составляющих небольшое множество состояний $𝒮	 = \{high, low\}$. В каждом состоянии агент может принять следующие решения: (1) продолжать активный поиск мусора в течение некоторого периода времени $(search)$; (2) оставаться на месте и ждать $(wait)$; (3) направиться к станции зарядки, чтобы подзарядить аккумулятор $(recharge)$. Если уровень заряда высокий $(high)$, то подзаряжаться бессмысленно, поэтому действие $recharge$ не включено в множество действий для этого состояния. Таким образом, имеем два множества действий: $𝒜(high) = \{search, wait\}$
и $𝒜(low) = \{search, wait, recharge\}$.

Вознаграждение в большинстве случаев равно нулю, но становится положительным, когда робот находит мусор, и большим отрицательным, если аккумулятор полностью разряжается. Если робот ведет активный поиск мусора, то существует возможность разряда аккумулятора. В таком случае робот остановится и будет ждать, пока кто-нибудь придет на помощь (получая при этом низкое вознаграждение). Если робот просто ждет ($wait$) на месте, то он находится в энергосберегающем режиме и посадить аккумулятор не может.

Если уровень заряда $high$, то уровень заряда с вероятностью $α$ останется $high$ и с вероятностью $1 – α$ станет $low$. Если уровень заряда $low$, то уровень заряда с вероятностью $β$ останется $low$ и с вероятностью $1 – β$  полностью разрядится. В последнем случае робота придется выручать с вознаграждением -3, после чего заряд вернется на уровень $high$. За каждый найденный мусор робот получает вознаграждение 1.

Обозначим $r_{search}$ и $r_{wait}$, где $r_{search}$ > $r_{wait}$, ожидаемое число мусора, собранное роботом во время поиска и во время ожидания соответственно. Наконец, предположим, что когда робот направляется к станции зарядки или стоит с разряженным аккумулятором, он не собирает мусор. Эта система представляет собой МППР, в таблице ниже выписаны вероятности переходов, ожидаемые вознаграждения и динамика.



|s|a|s'| вероятность $p(s'/s,a)$|ожидаемое $r(s,a,s')$|
|-|:-|:-|:-|:-|
|high|search|high|$α$|$r_{search}$
|high|search|low|$1-α$|$r_{search}$
|low|search|high|$1-β$|-3
|low|search|low|$β$|$r_{search}$
|high|wait|high|1|$r_{wait}$
|high|wait|low|0|-
|low|wait|high|0|-
|low|wait|low|1|$r_{wait}$
|low|recharge|high|1|0
|low|recharge|low|0|-

Отметим, что в таблице присутствуют строки для всех возможных комбинаций текущего состояния $s$, действия $a$ и следующего состояния $s'$.

Полезный способ визуализации динамики МППР дает граф переходов, здесь отображены еще и действия, в отличие от изображений марковских цепей:

image.png

В нем имеются вершины двух видов: **<font color='orange'>вершины состояний</font>** и **<font color='orange'>вершины действий</font>**. Для каждого возможного состояния присутствует вершина состояния (большой белый кружок, помеченный именем состояния), а для каждой пары состояние–действие – вершина действия (маленький черный кружок, помеченный именем действия и соединенный линией с вершиной состояния).

## Доход и эпизоды

image.png

Теперь более подробно рассмотрим, как агент может оценить состояние, в которое он собирается перейти. Так как цель агента – максимизировать полное вознаграждение, полученное за длительное время, а состояний и переходов в среде теперь больше одного, то нужны дополнительные расчеты. Для этого вводится в рассмотрение еще одно понятие - **<font color='orange'>ожидаемый доход</font>**, обозначаемый $G_t$,
определенный как некоторая функция от последовательности вознаграждений.

(5) $$G_t = R_{t+1} + γR_{t+2} + γ^2R_{t+3} + ... + γ^{T-1}R_{T} = \sum_{k=t+1}^{T}γ^{k-t-1}R_{k},$$
$$или$$
$$G_t = R_{t+1} + γR_{t+2} + γ^2R_{t+3} + ... + γ^{T-1}R_{T} = R_{t+1} + γ(R_{t+2} + γR_{t+3} + γ^2R_{t+4} + ...) = R_{t+1} + γG_{t+1},$$

где $T$ – terminal, **<font color='orange'>последний временной шаг</font>**, а $γ$ - **<font color='orange'>коэффициент обесценивания</font>**, $0\leq γ \leq 1$.

Когда взаимодействие агента с окружающей средой естественным образом распадается на подпоследовательности, называемые эпизодами, например партии в игре, обходы лабиринта или любое другое повторяющееся взаимодействие, то такие задачи называют **<font color='orange'>эпизодическими</font>**, а $T$ – определено числом ходов агента.

Если задача не распадается на эпизоды, как, например, задача управления роботом-пылесосом, то это **<font color='orange'>непрерывная</font>** задача, а $T = \infty$.

Коэффициент обесценивания определяет ценность будущих вознаграждений с точки зрения настоящего: вознаграждение, которое будет получено спустя $k$ временных шагов, стоит в $γ^{k–1}$ раз меньше, чем если бы оно было получено прямо сейчас.


Чтобы равенство выше было справедливо для всех временных шагов $t < T$, даже если завершение имеет место в момент $t + 1$, следует положить $G_T = 0$. Что справедливо, ведь после достижения терминального состояния наград больше не будет.

Следует отметить, что при $T = \infty$, $G_t$ становится суммой бесконечных слагаемых. Но в этом случае применяется коэффициент обесценивания $γ < 1$ и при постоянном ненулевом вознаграждении доход все же будет конечен и его можно будет оценить. И агент будет выбирать действия так, чтобы сумма обесцененных вознаграждений, которые он получит в будущем, принимала максимальное значение.

Например, при $R = const = 1$ доход будет: $G_t = \sum_{k=1}^{∞}{γ^{k}} = \frac{1}{(1-γ)}$.

## Стратегии и функции ценности
>Почти все алгоритмы обучения с подкреплением включают функции ценности – функции состояний (или пар состояние–действие), которые оценивают, насколько хорошо для агента оказаться в данном состоянии (или насколько хорошо предпринять данное действие в данном состоянии). Разумеется, вознаграждения, которых агент может ожидать в будущем, зависят от предпринимаемых им действий. Поэтому и функции ценности определены по отношению к конкретным способам действий – политикам(стратегиям).

На основании ожидаемого дохода можно оценить разные действия и состояния. Разные функции оценки позволяют с опорой на стратегию сделать это по-разному. Начнем с определения стратегии.

Формально **<font color='orange'>стратегией</font>** называется отображение состояний на вероятности выбора каждого возможного действия. Если агент следует стратегии $π$ в момент $t$, то $π(a | s)$ – вероятность того, что $A_t = a$, если $S_t = s$. Методы обучения с подкреплением описывают, как изменяется стратегия агента в результате приобретения им опыта.

(6) $$\mathbb{E}[R_{t+1}|S_t=s] = \sum_{a}π(a|s)\sum_{s',r}p(s',r|s,a)r = \sum_{a}π(a|s)r(s,a)$$

Для табличных методов, которые мы рассматриваем, $π$ и есть табличка, в которой информация, как себя вести в каком либо состоянии. Как конкретно это все будет выглядеть при расчетах, посмотрим чуть позже.

**<font color='orange'>Функция ценности состояния</font>** $s$ при стратегии $π$, обозначаемая $v_π(s)$, – это ожидаемый доход, когда агент начинает работу в состоянии $s$ и в дальнейшем следует стратегии $π$.

(7) $$v_π(s) = \mathbb{E}_π[G_t|S_t=s] = \mathbb{E}_π[\sum_{k=0}^{\infty}γ^{k}R_{t+k+1}|S_t=s] = \sum_{a}π(a|s)q_π(s,a)$$

Отметим, что ценность заключительного состояния, если оно существует, всегда равна 0. Мы называем функцию $v_π$ **<font color='orange'>функцией ценности состояний при стратегии $π$</font>**.

**<font color='orange'>Ценность выполнения действия</font>** $a$ в состоянии $s$ при стратегии $π$, обозначаемая $q_π(s, a)$, определяется как ожидаемый доход, когда агент начинает работу в состоянии $s$, предпринимает действие $a$ и затем следует стратегии $π$.

(8)$$q_π(s,a) = \mathbb{E}_π[G_t|S_t=s,A_t=a] = \mathbb{E}_π[\sum_{k=0}^{\infty}γ^{k}R_{t+k+1}|S_t=s,A_t=a] = \sum_{s',r}p(s',r|s,a)[r+γv_π(s')]$$

Мы называем $q_π$ **<font color='orange'>функцией ценности действий</font>** при стратегии $π$.

Функции ценности $v_π$ и $q_π$ можно оценить из опыта. Например, если агент следует стратегии $π$ и хранит для каждого встретившегося состояния среднее фактических доходов, последовавших за этим состоянием, то это среднее будет сходиться к истинной ценности состояния $v_π(s)$, когда количество появлений этого состояния стремится к бесконечности. Если для каждого действия, предпринятого в этом состоянии, хранятся отдельные средние, то эти средние точно так же будут сходиться к ценностям действий, $q_π(s,a)$.

**<font color='orange'>Фундаментальное свойство функций ценности</font>** - удовлетворение рекуррентным соотношениям (как с формулой 5):

(9)$$v_π(s) = \mathbb{E}_π[G_t|S_t=s] = \mathbb{E}_π[\sum_{k=0}^{\infty}γ^{k}R_{t+k+1}|S_t=s] = \mathbb{E}_π[R_{t+1}+γG_{t+1}|S_t=s] = $$
$$ = \sum_{a}π(a|s)\sum_{s',r}p(s',r|s,a)\Big[r+γ\mathbb{E}_π[G_{t+1}|S_{t+1}=s']\Big] = $$
$$ = \sum_{a}π(a|s)\sum_{s',r}p(s',r|s,a)[r+γv_π(s')] = \sum_{a}π(a|s)q_π(s,a)$$

Это **<font color='orange'>уравнение Беллмана для $v_π(s)$</font>**. Оно выражает соотношение между ценностью состояния $s$ и ценностями следующих за ним состояний.

(10)$$q_π(s,a) = \mathbb{E}_π[G_t|S_t=s, A_t=a] = \mathbb{E}_π[R_{t+1}+γG_{t+1}|S_t=s, A_t=a] = \sum_{s',r}p(s',r|s,a)[r+γ\sum_{a'}π(a'|s')q_π(s',a')]$$

Это **<font color='orange'>уравнение Беллмана для $q_π(s,a)$</font>**.

Дополнительно есть еще одна оценка - функция преимущества действия или функция преимущества, A-функция: $a_π(s, a) = q_π(s, a) - v_π(s)$. Она показывает разницу между функцией ценности действия $a$ в состоянии $s$ и функцией ценности состояния $s$ при следовании $π$.

В течение изучения простых табличных методов мы не будет прибегать к ее использованию, она пригодиться в более сложном курсе глубокого обучения с подкреплением. На практике она показывает существование разницы между двумя основными функциями ценности $q_π(s, a)$ и $v_π(s)$.

## Оптимальные стратегии и оптимальные функции ценности
>Решение задачи обучения с подкреплением означает, грубо говоря, нахождение стратегии, которая дает наибольшее вознаграждение за длительный период времени.

Теперь необходимо понять, как определить, какая стратегия игры лучше.

Стратегия $π$ больше (лучше) или равна стратегии $π'$, тогда и только тогда, когда $v_π(s) \leq v_{π'}(s)$ для всех $s ∈ 𝒮$. Всегда существует хотя бы одна стратегия, большая или равная всем остальным стратегиям. Это **<font color='orange'>оптимальная стратегия</font>**. Хотя оптимальных стратегий может быть несколько, мы обозначаем любую из них $π^*$. У них у всех одна и та же функция ценности состояний, которая называется оптимальной функцией ценности состояний, обозначается $v*$ и определяется следующим образом:

$v^*(s) = max_π v_π(s)$ $\forall s∈\mathscr{S}$.

У оптимальных стратегий также одна и та же функция ценности действий, которая обозначается $q^*$ и определяется следующим образом:

$q^*(s,a) = max_π q_π(s,a)$ $\forall s∈\mathscr{S}, a∈\mathscr{A}(s)$.

Для пары $(s,a)$ эта функция дает ожидаемый доход от выбора действия $a$ в состоянии $s$ и далее следования оптимальной стратегии. Таким образом, мы можем следующим образом выразить $q^*$ через $v^*$:

$q^*(s,a) = \mathbb{E}[R_{t+1}+γv^*(S_{t+1})|S_t=s,A_t=a]$


Уравнение Беллмана для $v^*$ или **<font color='orange'>уравнение оптимальности Беллмана</font>** интуитивно выражает тот факт, что ценность состояния при оптимальной стратегии должна быть равна ожидаемому доходу для лучшего действия в данном состоянии:

(11)$$v^*(s) = max_{a∈\mathscr{A}(s)} q_{π^*}(s,a) = max_{a}\mathbb{E}_{π^*}[G_t|S_t=s,A_t=a] = max_{a}\mathbb{E}_{π^*}[R_{t+1}+γG_{t+1}|S_t=s,A_t=a] = max_{a}\mathbb{E}[R_{t+1}+γv^*(S_{t+1})|S_t=s,A_t=a] = max_{a}\sum_{s',r}p(s',r|s,a)[r+γv^*(s')]$$

Последние два равенства – это две формы уравнения оптимальности Беллмана для $v^*$.

**<font color='orange'>Уравнение оптимальности Беллмана для $q^*$ </font>**имеет вид:

(12)$$q^*(s,a) = \mathbb{E}[R_{t+1}+γmax_{a'}q^*(S_{t+1},a')|S_t=s,A_t=a] = \sum_{s',r}p(s',r|s,a)[r+γmax_{a'}q^*(s',a')]$$

В диаграммах выше дуга означает выбор максимума по вариантам выбора.

Для конечных МППР уравнение оптимальности Беллмана (11) имеет единственное решение $v^*$, не зависящее от стратегии. В действительности уравнение оптимальности Беллмана – это **<font color='orange'>система уравнений</font>**, по одному для каждого состояния, так что при наличии $n$ состояний мы имеем систему из $n$ уравнений с $n$ неизвестными. Если динамика $p$ окружающей среды известна, то в принципе эту систему уравнений можно решить относительно $v^*$, применяя любой из многочисленных методов решения систем нелинейных уравнений. Тоже самое для $q^*$.

**<font color='orange'>Любая стратегия, жадная относительно оптимальной функции ценности $v^*$, является оптимальной.</font>** Термин «жадный» в данном случае описывает стратегии, которые выбирают действия, учитывая только краткосрочные последствия, т.е. такая стратегия будет выбирать только действия с максимальной оценкой.

## Применение на практике
На практике почти во всех задачах через Беллмана решить все **<font color='orange'>невозможно</font>**, т.к. нужно:

1.   точно знать динамику среды;
2.   достаточно вычислительных ресурсов для полного вычисления всех шагов в среде;
3.   чтобы имело место марковское свойство.

Доступная память – также важное ограничение. Большой объем памяти часто необходим для аппроксимации функций ценности, стратегий и моделей. В задачах с небольшим конечным множеством состояний такие аппроксимации можно построить в  виде массивов или таблиц, в  которых имеется одна запись для каждого состояния (или пары состояние–действие). Мы называем это **<font color='orange'>табличным случаем</font>**, а соответствующие методы решения – **<font color='orange'>табличными методами</font>**. Но во многих случаях,  представляющих практический интерес, состояний так много, что никакой таблицы не хватит. Тогда функции необходимо **<font color='orange'>аппроксимировать с помощью какого-то более компактного представления в виде параметрической функции</font>**. Таким образом, чаще всего нам придется идти на компромисс.

Благодаря онлайновой природе обучения с подкреплением можно аппроксимировать оптимальные стратегии, так чтобы больше усилий тратилось на обучение умению принимать хорошие решения в часто встречающихся состояниях ценой меньшего внимания к редко встречающимся. Это то ключевое свойство, которое отличает обучение с подкреплением от других подходов к приближенному решению МППР.

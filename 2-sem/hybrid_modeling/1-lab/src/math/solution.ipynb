```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python
N_0 = 5
r = 0.05   # коэффициент роста (alpha - beta s.t alpha > beta )
t_0 = 0    # начальное время
t = np.linspace(0, 400, 100)
k = 40
```

```python
# Уравнение
N = N_0 * np.exp(r * (t - t_0))

# Построение графика
plt.plot(t, N)
plt.xlabel('Время')
plt.ylabel('N')
plt.title('Экспоненциальный рост')
plt.scatter(t_0, N_0, color='red', label=f'Момент времени $t_0={t_0}$')
plt.grid(True)
plt.show()
```

```python

# partial solution of the equation
N = ( k * N_0 * np.exp(r * (t - t_0)) ) / ( k + N_0 * ( np.exp(r * (t - t_0)) - 1 ))


plt.plot(t, N)
plt.xlabel('Время')
plt.ylabel('N')
plt.scatter(t_0, N_0, color='red', label=f'Момент времени $t_0={t_0}$')
plt.grid(True)
plt.show()

```

```python
C = 1 # resource Growth rate
b = 0.01 # resource Consumption rate
```

```python
from scipy.integrate import odeint

# Функция правой части системы
def f( y, t):
  [N, F] = y
  return [r * N * (1 - N / (k * F)), C - b * N]

F_0 = 100

# Численное решение
y = odeint(f, [N_0, F_0], t)

plt.plot(t, y[:, 0], label="N(t) Численность")
plt.plot(t, y[:, 1], label="F(t) Ресурс")
plt.xlabel("Время")
# plt.grid()
plt.legend()
plt.show()
```
